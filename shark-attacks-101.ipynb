```python
# amount of unique values for each column
unique_counts = df.nunique()
print("\nNumber of unique values for each column:")
print(unique_counts)



#code for different counts like top 5 attacks, top 5 sports, etc




sports_count = df["Activity"].value_counts()
top_five_sports = {activity: count for activity, count in sports_count.nlargest(5).items()}
print(f"The top 5 activities are: \n{sports_count}")

atk_count = df["Country"].value_counts()
top_five_most = {country: count for country, count in atk_count.nlargest(5).items()}
top_five_less = {country: count for country, count in atk_count.nsmallest(5).items()}
print(f"The top 5 locations are: \nmost common: {top_five_most} \nless common: {top_five_less}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the count plot
sns.countplot(data=shark_df, x='Country', order=shark_df['Country'].value_counts().index[:20])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set title and labels
plt.title('Shark Attacks by Country', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Show the plot
# plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
